```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_image(img1, img2):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(img1)
    axes[0].set_title("Original Image")
    axes[0].axis("off")

    axes[1].imshow(img2)
    axes[1].set_title("Image with Median Filter")
    axes[1].axis("off")

    plt.show()
```

The image has a lot of salt and pepper noise, i.e. sparsely occurring black and white pixels. Then the best choice is median filtering since 

  - It will take the median value in any neighborhood of pixels, so;  
  - The *sparse* noise pixels will thus be attenuated.  

This choice is more appropriate than some other candidates:  

- Gaussian filtering would take into account these tiny noise details in its blurring effect, which is less ideal.  
  - There are also no small details which we must preserve.
- Similarly, uniform filtering will factor in the noise pixels when computing the mean, thus leading to a loss of detail.  

```python
image = cv2.imread("images/Task2.jpg")

denoised_image = cv2.medianBlur(image, 5)

show_image(image, denoised_image)
```
